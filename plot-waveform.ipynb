```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile

# Load the WAV file
sample_rate, audio = wavfile.read("low_pass_drums.wav")

# Define the time range in seconds
start_time = 11	# Start time in seconds
end_time = 13	# End time in seconds

# Calculate the corresponding sample indices
start_index = int(start_time * sample_rate)
end_index = int(end_time * sample_rate)

# Extract the audio segment for the specified time range
x = np.arange(start_time, end_time, 1/sample_rate)
y = audio[start_index:end_index]

# Normalize the audio data to the range [-1.0, 1.0]
y = y / np.max(np.abs(y))

# Create the plot
plt.plot(x, y)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Audio Waveform')

# Show the plot
plt.show()
```
