```python
import matplotlib.pyplot as plt
from matplotlib.axes import Axes
import soundfile as sf
from autostep3.util import cutoff_samples, stereo2mono

AUDIO_FILE = "separated_audio/obsessed/drums.mp3"
CUTOFF_THRESHOLD = 0.5

audio, samplerate = sf.read(AUDIO_FILE)
audio = stereo2mono(audio)

# Get only audio between 15 and 30 seconds
audio_15_30 = audio[15*samplerate:30*samplerate]

cutoff_samples(audio_15_30, 0.5)

ax: Axes
fig, ax = plt.subplots(1, figsize=(20, 5))
ax.plot(audio_15_30)
```
