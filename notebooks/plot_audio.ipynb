```python
import matplotlib.pyplot as plt
import soundfile as sf
import numpy as np

SONG_PATH = "test_audio/hyperpop.mp3"
SECONDS_RANGE = (0, 0.1)

audio, samplerate = sf.read(SONG_PATH)
audio = audio[int(SECONDS_RANGE[0] * samplerate) : int(SECONDS_RANGE[1] * samplerate)]
fig, ax = plt.subplots(1, figsize=(20, 5))
ax.plot(np.arange(len(audio)), audio)
fig.show(False)
```
