```python
from song import Song
import numpy as np
import matplotlib.pyplot as plt
import code

start_sec = 16
end_sec = 28

song = Song('../test_audio/vocals.wav', bpm = 128, offset = -0.095)
song.removeQuiet(0.25)

start_sample = start_sec * song.samplerate
end_sample = end_sec * song.samplerate
num_samples = end_sample - start_sample
x = np.linspace(start_sample, end_sample, num_samples)
y = song.data[start_sample : end_sample]

xs = x / song.samplerate

fig, ax = plt.subplots()
ax.plot(xs, y)
plt.show()
```
